```python
import pickle as pkl
import seaborn as sns; sns.set()
import matplotlib as plt
import pandas as pd

import plotly.express as px
sns.set(rc={'figure.figsize':(15, 6)})
```

```python
def make_df(file):
    with open(file,'rb') as fp:
        data = pkl.load(fp)
    df = pd.DataFrame(columns=['batch num.', 'phi_val', 'delta', 'lt_thresh'])
    
    df['batch num.'] = list(data.keys())
    df['phi_val'] = [i[0] for i in data.values()]
    df['delta'] = [i[1] for i in data.values()]
    df['lt_thresh'] = [i[2] for i in data.values()]
    return df
    
```

## SimpleConvNet Results

```python
convnet_p1_df = make_df('new-simple-conv-data-p1.pkl')
```

```python
convnet_p1_ax = sns.scatterplot(data=convnet_p1_df, x="batch num.", y="phi_val", hue='lt_thresh', size='delta')
```

```python
convnet_p1_delta_ax = sns.scatterplot(data=convnet_p1_df, x="batch num.", y="delta", hue='lt_thresh')
```
